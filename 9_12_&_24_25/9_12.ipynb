```python
import pandas as pd

# Load the data
iris = pd.read_csv('iris.csv')

# Display first few rows
iris.head()

```

```python
# Create subsets
setosa = iris[iris['variety'] == 'setosa']
versicolor = iris[iris['variety'] == 'versicolor']
virginica = iris[iris['variety'] == 'virginica']

# Check sizes
print('Setosa:', setosa.shape)
print('Versicolor:', versicolor.shape)
print('Virginica:', virginica.shape)
```

```python
# Merge setosa and versicolor
merged = pd.concat([setosa, versicolor], axis=0)

# Check merged result
merged.head()
```

```python
# Sort by petal_length
sorted_iris = iris.sort_values(by='petal.length')

# Show top 5 sorted rows
sorted_iris.head()
```

```python
# Transpose the DataFrame
transposed = iris.T

# Display transposed data
transposed.head()
```

```python
# Melt the data (wide to long)
melted = pd.melt(iris, id_vars=['variety'], var_name='measurement', value_name='value')

# Show melted data
melted.head()
```

```python
# Pivot (cast) the melted data back to wide format
wide = melted.pivot_table(index=iris.index, columns='measurement', values='value')

# Reset index to make it clean
wide = wide.reset_index(drop=True)

# Show wide data
wide.head()
```
