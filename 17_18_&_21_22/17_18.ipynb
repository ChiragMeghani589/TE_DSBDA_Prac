```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('heart.csv')

# Preview the dataset
print(df.head())


```

```python
# ---------- a. Histograms ----------
# Objective: Understand the distribution of numerical features like age, cholesterol, etc.

df.hist(column=['age', 'chol', 'trestbps'], bins=20, figsize=(12, 6))
plt.suptitle('Histogram of Age, Cholesterol, and Resting BP')
plt.show()

```

```python
# ---------- b. Pie Charts ----------
# Objective: Visualize the proportion of heart disease presence (target variable)

if 'target' in df.columns:
    target_counts = df['target'].value_counts()
    plt.figure(figsize=(6, 6))
    plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title('Pie Chart of Heart Disease Presence (target)')
    plt.show()
else:
    print("Column 'target' not found. Please check the dataset.")

```

```python
# ---------- c. Box Plots ----------
# Objective: Detect outliers and compare distributions between heart disease groups

if 'target' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='target', y='age', data=df)
    plt.title('Boxplot of Age by Heart Disease Presence')
    plt.show()
else:
    print("Column 'target' not found. Please check the dataset.")

```

```python
# ---------- d. Scatter Plots ----------
# Objective: Check relationship between age and cholesterol with disease presence

if 'target' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='age', y='chol', hue='target', data=df, palette='Set1')
    plt.title('Scatter Plot of Age vs Cholesterol by Heart Disease Presence')
    plt.show()
else:
    print("Column 'target' not found. Please check the dataset.")

```

```python
# ---------- e. Add boxplots to a scatterplot ----------
# Objective: Overlay boxplots on scatterplot to visualize spread and distribution together

if 'target' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='target', y='chol', data=df, whis=[0, 100], width=0.4, palette='Pastel1')
    sns.stripplot(x='target', y='chol', data=df, jitter=True, color='black', alpha=0.5)
    plt.title('Scatter + Boxplot of Cholesterol by Heart Disease Presence')
    plt.show()
else:
    print("Column 'target' not found. Please check the dataset.")
```

```python
# ---------- Combined Histogram + Boxplot ----------
# Objective: Show data distribution (histogram) + summary stats (boxplot) together

feature = 'age'  # you can change this to 'chol', 'trestbps', etc.

fig, axes = plt.subplots(2, 1, figsize=(10, 8), gridspec_kw={'height_ratios': [1, 4]})

# Boxplot on top
sns.boxplot(x=df[feature], ax=axes[0], color='lightblue')
axes[0].set(xlabel='')

# Histogram below
sns.histplot(df[feature], bins=30, ax=axes[1], color='skyblue', kde=True)
axes[1].set_xlabel(f'{feature.capitalize()}')
axes[0].set_title(f'Combined Boxplot + Histogram of {feature.capitalize()}')

plt.tight_layout()
plt.show()


# ---------- Combined Scatter Plot + Boxplot ----------
# Objective: Overlay scatter points on boxplot to see spread + outliers

if 'target' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='target', y='chol', data=df, whis=[0, 100], width=0.5, palette='Pastel1')
    sns.stripplot(x='target', y='chol', data=df, jitter=True, color='black', alpha=0.5)
    plt.title('Combined Boxplot + Scatter Plot of Cholesterol by Heart Disease Presence')
    plt.xlabel('Heart Disease Presence (0 = No, 1 = Yes)')
    plt.ylabel('Cholesterol')
    plt.show()
else:
    print("Column 'target' not found. Please check the dataset.")
```
